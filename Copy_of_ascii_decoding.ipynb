# ASCII Encoding and Decoding Practice

For reference, here is the ASCII chart (codes 0 to 31 are control codes; code 32 is the space character).

image.png

Note that Python has built-in functions `ord` and `chr` for converting single codes to characters and back.  We can use these functions for encoding strings into a list of codes and vice-versa.

```python
help(ord)
```

```python
help(chr)
```

```python
ord('O')
```

```python
chr(79)
```

```python
# convert a string to a list of codes
def encode(string):
  return [ord(char) for char in string]

# convert a list of codes to a string
def decode(codes):
  return ''.join(chr(c) for c in codes)

codes = encode("Bello")
print('codes:', codes)
decoded = decode(codes)
print('codes decoded:', decoded)
```
